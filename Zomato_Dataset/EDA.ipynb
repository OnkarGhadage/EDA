Importing the another dataset which have country codes.

```python
df_cc = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_cc
```

So as we can see there are total 15 countries.  
Each country has its unique country code.

---
So now we have country code and its corresponding names  
Now we have to combine them or merge them to know in our df which country is that.

```python
final_df = pd.merge(df, df_cc, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

Now we have added a column i.e. country.  

`on='Country code'` - It mean on which column we want to merge two df's.  
`how='left'` - Left join.  

Its just a normal SQL type of concept of left join. We can also do inner join but oour data is in left side so we just focused in that.

---
Lets just explore the country names and counts

```python
final_df.Country.value_counts()
```

We store the country names and corresponding value in 2 variables. And it is a dataframe so we can do it like.  

```python
country_name = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
country_name
```

```python
country_val
```

```python
# Now lets plot it in pie chart
plt.pie(country_val, labels=country_name)
plt.show()
```

```python
# Looking bad lets just do some changes
# Top 3 countries
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')    # Added percentages '%1.2f%%' mean 2 decimal places
plt.show()
```

### Observations
From above activities we can know that the dataset have maximum records from INDIA and on 2nd place we have US and then UK and so on.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

### Observations
1. Rating 4.5 to 4.9 -> Excellent
2. Rating 4.0 to 4.4 -> Very Good
3. Rating 3.5 to 3.9 -> Good
4. Rating 2.5 to 3.4 -> Average
5. Rating ... to 2.4 -> Poor
6. Not rated -> 0.0

```python
ratings.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=ratings, x='Aggregate rating', y='Rating Count', hue='Rating color', palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
plt.tight_layout()
plt.show()
```

```python
ratings['Rating Count'][1:].sum()
```

### Observation
So from above graph we can know that maximum ratings are given have orange color.  
Total 7403 users rated and 2148 not rated.
