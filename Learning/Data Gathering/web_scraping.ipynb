# web scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

**Filter out all h1 tags from the webpage**

```python
soup.find_all('h1')
```

```python
soup.find_all('h1')[0]
```

```python
soup.find_all('h1')[0].text
```

**Now h2 tags**

```python
soup.find_all('h2')
```

```python
len(soup.find_all('h2'))
```

```python
soup.find_all('h2')[0].text
```

```python
soup.find_all('h2')[0].text.strip()
```

```python
for i in soup.find_all('h2')[1:-3]:
    print(i.text.strip())
```

**Now lets just extract rating of companies**

```python
len(soup.find_all('div', class_='rating_text rating_text--md'))
```

```python
for i in soup.find_all('div', class_='rating_text rating_text--md'):
    print(i.text.strip())
```

**Now number of reviews**

```python
len(soup.find_all('span', class_='companyCardWrapper__companyRatingCount'))
```
