## Trying to get name, weight and height of all pokemons

```python
data = requests.get('https://pokeapi.co/api/v2/pokemon/1')
pd.DataFrame([data.json()])[['name', 'weight', 'height']]
```

```python
df = pd.DataFrame()
for i in range(1, 1351):
    data = requests.get(f"https://pokeapi.co/api/v2/pokemon/{i}")
    temp = pd.DataFrame([data.json()])[['name', 'weight', 'height']]
    df = pd.concat([df, temp], ignore_index=True)

df
```
