# Data Gathering

## Reading the data

### From CSV

CSV - Comma separated first  
The values are separated by a comma

```python
pd.read_csv("file_name.csv")        # If file is in the same folder
pd.read_csv("path/to/file.csv")     # If file is in another folder
```

#### Parameter:

1. `sep=".."` - The files might have different seperator like " "(space), "    "(tab), etc. So we can change it.
2. `names=[...]` - If file does not contain column names then we can pass list of column names.
3. `index_col='col_name'` - If want a column as index then pass that name in this.
4. `header=1` - The csv file have header but the row is other than 0(deafult).
5. `use_col=[col_name(s)]` - Pass list of column names that you want to extract from csv file. Only that columns are readed.
6. `skip_rows=[1,2,3,..]` - If want to ignore some rows from csv file then pass the rows no. 0 based indexing
7. `nrows=100` - It only take first 100 rows.
8. `encoding='UTF-8'`
9. `error_bad_lines=False` - It ignores bad lines
10. `dtype={'col_name':date,....}` - change the data type of columns
11. `na_values=[' ', '$', ...]` - All cells having these values are considered as nan values
